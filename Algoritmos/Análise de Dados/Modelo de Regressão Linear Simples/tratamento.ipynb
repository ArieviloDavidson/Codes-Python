```python
import matplotlib.pyplot as plt
import pandas as pd 
import math
# import xlsxwriter
```

```python
dados_csv = pd.read_csv("population_by_country_2020.csv")
y_csv = dados_csv["Population (2020)"]
x_csv = dados_csv["Land Area (KmÂ²)"]

x = []
y = []

for i in range(x_csv.shape[0]):
    x.append(x_csv[i])
    y.append(y_csv[i])
```

```python
Sx = 0
Sy = 0
Sxy = 0
Sxx = 0
Syy = 0
x_x_med = []
y_y_med = []
xx = []
yy = []
xy = []

for i in range(len(x)):
    Sx += x[i]
    Sy += y[i]
x_med = Sx/len(x)
y_med = Sy/len(y)

for i in range(len(x)):
    x_x_med.append((x[i] - x_med))
    y_y_med.append((y[i] - y_med))

for i in range(len(x)):
    xx.append(x_x_med[i]**2)
    yy.append(y_y_med[i]**2)
    xy.append(y_y_med[i]*x_x_med[i])

# Calculo do Rho
for i in range(len(x)):
    Sxy += xy[i]
    Sxx += xx[i]
    Syy += yy[i]

rho = Sxy/ math.sqrt(Sxx*Syy)
print(rho)
```

```python
B1 = Sxy/Sxx
B0 =y_med - B1*x_med
sigma = Syy/len(x)

print("A reta Ã© -> y=",B0," + ",B1,"x")
```

```python
new_y = []
for i in range(len(x)):
    new_y.append(B0 + B1*x[i])

plt.plot(x,new_y)
plt.scatter(x,y)
plt.show()
```
