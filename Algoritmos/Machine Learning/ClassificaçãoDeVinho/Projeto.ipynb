```python
# Carregando o conjunto de dados
import pandas as pd
arquivo = pd.read_csv('./dados_vinhos.csv')
arquivo.shape
```

```python
arquivo.head() # Para visualizar o conjunto de dados no python (5 primeiras linhas padrão)
```

```python
arquivo['style'] = arquivo['style'].replace('red', 0) # Transforma ou "troca" valores da coluna descrita
arquivo['style'] = arquivo['style'].replace('white', 1) # Transforma ou "troca" valores da coluna descrita
```

```python
tipo = arquivo['style'] # Variável alvo (o que quero prever)
dados = arquivo.drop('style', axis = 1) # Restante dos dados
```

```python
from sklearn.model_selection import train_test_split

# Cria conjustos para treino e também para testes
x_treino, x_teste, y_treino, y_teste = train_test_split(dados, tipo, test_size = 0.3)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
# Criando o modelo
modelo = ExtraTreesClassifier()
modelo.fit(x_treino,y_treino)

resultado = modelo.score(x_teste,y_teste) # Vai comparar os dados de treino com os reais
print("Acurácia:", resultado)
```
